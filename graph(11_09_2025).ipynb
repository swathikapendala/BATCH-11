```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.plot(x, y, color='blue', linestyle='--', marker='o', label='y=2x')
plt.title("Line Chart Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
categories = ['A','B','C','D']
values = [12,17,20,8]
plt.bar(categories, values, color='orange', width=0.5)
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
x = [5,7,8,7,6,9,5,6]
y = [99,86,87,88,100,86,103,87]
plt.scatter(x, y, color='red', s=100, alpha=0.7)
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
sizes = [20, 30, 25, 25]
labels = ['vhyshu', 'Bunny', 'Chinnu', 'dolly']
colors = ['gold','lightblue','lightgreen','pink']
explode = [0,0.1,0,0]  # Highlight Bananas
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, explode=explode)
plt.title("Pie Chart Example")
plt.show()
```

```python
import seaborn as sns
import numpy as np
data = np.random.randn(1000)  # Random normal distribution
sns.histplot(data, bins=30, kde=True, color='purple')
plt.title("Distribution Plot Example")
plt.show()
```

```python
data = [7,8,5,6,9,12,15,14,5,6,7,20]
sns.boxplot(data=data)
plt.title("Box Plot Example")
plt.show()
```

```python
iris = sns.load_dataset("iris")
sns.violinplot(x="species", y="sepal_length", data=iris)
plt.title("Violin Plot Example")
plt.show()
```

```python
sns.pairplot(iris, hue="species")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y1 = [2,4,6,8,10]
y2 = [1,3,5,7,9]
plt.plot(x, y1, color='blue', label="y=2x")
plt.plot(x, y2, color='red', linestyle='--', label="y=2x-1")
plt.title("Line Chart with Labels")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()  # show labels
plt.show()
```

```python
categories = ['A','B','C','D']
values1 = [10,15,20,25]
values2 = [8,18,22,20]
plt.bar(categories, values1, label="2024", color="green")
plt.bar(categories, values2, alpha=0.6, label="2025", color="orange")
plt.title("Bar Chart with Labels")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.figure(figsize=(8,5))
plt.plot(x, y,
         color='purple',
         linestyle='--',
         linewidth=2,
         marker='o',
         markersize=10,
         markerfacecolor='yellow',
         markeredgecolor='black',
         alpha=0.8,
         label="y=2x")
plt.title("Customized Line Chart", fontsize=14, color='blue')
plt.xlabel("X-axis", fontsize=12)
plt.ylabel("Y-axis", fontsize=12)
plt.grid(True, linestyle=':')
plt.legend(loc='upper left')
plt.xlim(0,6)
plt.ylim(0,12)
plt.show()
```

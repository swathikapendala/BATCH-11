# Task
Load the titanic dataset from seaborn, analyze it by displaying the first few rows, checking the data types, and looking for missing values, and visualize the data by plotting the distribution of age, fare, and survival status.

## Load data

### Subtask:
Load the titanic dataset from seaborn.


**Reasoning**:
Import seaborn and load the titanic dataset as instructed.



```python
import seaborn as sns

titanic_df = sns.load_dataset('titanic')
```

## Analyze data

### Subtask:
Analyze the data by displaying the first few rows, checking the data types, and looking for missing values.


**Reasoning**:
Display the first 5 rows of the dataframe, check the data types, and calculate the number of missing values for each column.



```python
display(titanic_df.head())
display(titanic_df.info())
display(titanic_df.isnull().sum())
```

## Visualize data

### Subtask:
Visualize the data by plotting the distribution of age, fare, and survival status.


**Reasoning**:
The subtask requires visualizing the distribution of 'age', 'fare', and 'survived' columns. Creating histograms for 'age' and 'fare' and a count plot for 'survived' will accomplish this. These plots will help understand the data distributions.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
sns.histplot(data=titanic_df, x='age', kde=True, bins=20)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 5))
sns.histplot(data=titanic_df, x='fare', kde=True, bins=20)
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8, 5))
sns.countplot(data=titanic_df, x='survived')
plt.title('Distribution of Survival Status')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Not Survived', 'Survived'])
plt.show()
```
